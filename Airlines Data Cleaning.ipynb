```python
# Converting the Data Types
data["id"] = data["id"].astype("object")
data["day"] = data["day"].astype("category")
data["dest_region"] = data["dest_region"].astype("category")
data["dest_size"] = data["dest_size"].astype("category")
```

```python
data[["dest_region"]]
```

```python
# Replacing the entry eur with Europe
data["dest_region"] = data["dest_region"].replace(["eur"], ["Europe"])
```

```python
data.head(50)
```

```python
# Renaming the boarding_area colum to boarding_gate
data.rename({"boarding_area":"boarding_gate"},axis=1, inplace=True)
```

```python
# Stripping the gates
data["boarding_gate"] = data["boarding_gate"].str.strip("Gates ")
```

```python
# Converting the Data Types
data["cleanliness"] = data["cleanliness"].astype("category")
data["safety"] = data["safety"].astype("category")
data["satisfaction"] = data["satisfaction"].astype("category")
data["dept_time"] = data["dept_time"].astype("datetime64")
```
